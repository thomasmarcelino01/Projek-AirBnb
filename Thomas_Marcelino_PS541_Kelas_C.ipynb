MEMBUAT TREND LINE TIAP PENDAPATAN TIAP BULAN DI NEIGHBOURHOOD

```python
dataset=pd.read_csv('/content/drive/MyDrive/DATA ANALISIS/data 1.csv')
```

```python
dataset['gmv'] = dataset['availability_365']*dataset['price']
```

```python
# Menambahkan kolom 'order_month' berdasarkan kolom 'date'
dataset['order_month'] = dataset['date'].str.slice(0,7)

# Menjalankan query SQL untuk menghitung pendapatan per bulan berdasarkan neighbourhood_group 'Central Region'
Pendapatan_per_bulan_neighbourhood = ps.sqldf("""
                            SELECT order_month, neighbourhood, SUM(gmv) AS gmv
                            FROM dataset
                            WHERE neighbourhood_group IN ('Central Region')
                            GROUP BY order_month, neighbourhood
                            HAVING neighbourhood IN (
                                SELECT neighbourhood
                                FROM dataset
                                WHERE neighbourhood_group = 'Central Region'
                                GROUP BY neighbourhood
                                ORDER BY SUM(gmv) DESC
                                LIMIT 3
                            )
                            ORDER BY order_month
                            """)

# Daftar urutan warna
color_sequence = ["red", "blue", "maroon", "#FF8400", "#7AA874"]

# Membuat plot line dengan Plotly Express
fig_line = px.line(Pendapatan_per_bulan_neighbourhood, x="order_month", y="gmv", color="neighbourhood",
                   template="simple_white", color_discrete_sequence=color_sequence, markers=True)

# Mengubah judul
fig_line.update_layout(title="<b>Top 3 Total Pendapatan Central Region Setiap Bulan</b>", title_font=dict(size=20, family="arial"))

# Mengubah label sumbu X & Y
fig_line.update_xaxes(title="<b>Order Month</b>", title_font=dict(size=13, family="arial"))
fig_line.update_yaxes(title="<b>Total Pendapatan (SGD)</b>", title_font=dict(size=13, family="arial"))

# Mengubah garis dan marker
fig_line.update_traces(line=dict(width=3), marker=dict(size=8))

# Menampilkan plot
fig_line.show()
```
